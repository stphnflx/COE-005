```python
# Import libraries to be use for processing the frames  
from gym.wrappers import GrayScaleObservation
# Import libraries for vectorization wrappers
from stable_baselines3.common.vec_env import VecFrameStack, DummyVecEnv
# Import Matplotlib for observation for frame stacking
from matplotlib import pyplot as plt
```

```python
#Setting up the environment of the game 

# Game environment creation
env = gym_super_mario_bros.make('SuperMarioBros-v0')
# Setting up inputs in the game
env = JoypadSpace(env, SIMPLE_MOVEMENT)
# Grayscaling for observation for the Reinforced Learning
env = GrayScaleObservation(env, keep_dim=True)
# Creating Dummy environment of the game
env = DummyVecEnv([lambda: env])
# Stacking the frames 
env = VecFrameStack(env, 4,channels_order="last")
```

```python
#reset environment
state=env.reset()
```

Training the Reinforced Learning Model

```python
# Importing library for file path management
import os 
# Import PPO for algos
from stable_baselines3 import PPO
# Import Base Callback for saving models
from stable_baselines3.common.callbacks import BaseCallback
```

```python
#Defining the class for saving the model
class TrainAndLoggingCallback(BaseCallback):

    def __init__(self, check_freq, save_path, verbose=1):
        super(TrainAndLoggingCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.save_path = save_path

    def _init_callback(self):
        if self.save_path is not None:
            os.makedirs(self.save_path, exist_ok=True)

    def _on_step(self):
        if self.n_calls % self.check_freq == 0:
            model_path = os.path.join(self.save_path, 'best_model_{}'.format(self.n_calls))
            self.model.save(model_path)

        return True
```

```python
#creating directory for checkpoint and logs 
CHECKPOINT_DIR = './train/'
LOG_DIR = './logs/'
```

```python
# Setup model saving callback
callback = TrainAndLoggingCallback(check_freq=10000, save_path=CHECKPOINT_DIR)
```

```python
# Defining parameters for the RL Model
model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=LOG_DIR, learning_rate=0.000001, 
            n_steps=512) 
```

```python
# Training the model
model.learn(total_timesteps=100000, callback=callback)
```

```python
#Saving the trained RL model 
model.save('thisisatestmodel')
```

```python
#run the game 
import gym
from gym import wrappers

env = gym_super_mario_bros.make('SuperMarioBros-v0')
env = wrappers.Monitor(env, "./gym-results", force=True)
env.reset()
for _ in range(3600):
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    if done: break
env.close()
```
