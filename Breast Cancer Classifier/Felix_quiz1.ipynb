```python
#reading and loading the dataset
import pandas as pd
dataset=pd.read_csv("data.csv")
print("done") #checking if the code is executed
```

```python
#display the first 5 rows of data
dataset.head()
```

```python
#storing data to convenient variable name
parameters=dataset.iloc[:,2:32] #using the parameters needed excluding diagnosis and ID number since it is not relevant 
print(parameters) #to check if the parameters included are correct
```

```python
diagnose=dataset["diagnosis"]  #storing the identification given the parameters
print(diagnose)
```

```python
#dividing the data into training and test data
from sklearn.model_selection import train_test_split
parameters_train, parameters_test, diagnose_train, diagnose_test = train_test_split(parameters, 
                                                                                    diagnose, 
                                                                                    test_size = 0.2,   #20% test size 80% training size
                                                                                    random_state = 100)
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(parameters_train, diagnose_train)

#Predict the response for test dataset
diagnose_pred = clf.predict(parameters_test)
print(diagnose_pred)
```

```python
#checking accuracy of prediction
from sklearn import metrics
# Model Accuracy Checking
print("Accuracy:",metrics.accuracy_score(diagnose_test, diagnose_pred))
```

```python
# Model Precision Checking
print("Precision:",metrics.precision_score(diagnose_test, diagnose_pred,pos_label='positive',average='weighted'))

#recall checking
print("Recall:",metrics.recall_score(diagnose_test, diagnose_pred,pos_label='positive',average='weighted'))
```

```python
#reading the test data
testdata=pd.read_csv("test.csv")
testdata.head()
```

```python
#predicting the test data
test_pred = clf.predict(testdata)
print(test_pred) #predicted diagnosis
```
